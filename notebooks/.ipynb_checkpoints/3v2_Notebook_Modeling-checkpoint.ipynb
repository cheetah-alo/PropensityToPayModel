```python
def plot_sampling(X, y, X_res, y_res, ax, fig = None, sampler = None, title=None):
    sns.set_style('whitegrid')
    sns.set_palette(SETCOL1)
    if fig:
        ax[0].scatter(X[:, 0], X[:, 1], c=y, alpha=0.8, edgecolor="g")
        ax[1].scatter(X_res[:, 0], X_res[:, 1], c=y_res, alpha=0.8, edgecolor="g")
        sns.despine(fig)
        ax[0].set_title(f"Before sampling")
        if sampler:
            ax[1].set_title(f"Resampling with {sampler.__class__.__name__}")
        else:
            ax[1].set_title("After sampling")
    else:
        print('pass the `fig` parameter')
```

####  <span style="color:#28BAE5"> Input Original  <a name="id3"></a>

```python
# set aside 30% of train and test data for evaluation
X_train, X_val, y_train, y_val = train_test_split(X, y,
    test_size=0.03, shuffle = True, random_state = 8, stratify = y)
```

####  <span style="color:#28BAE5"> Oversampling  <a name="id3"></a>
    
There are several methods available, but I am going to test some of them, and the first one that I decided to apply is the most common data augmentation technique is known as Synthetic Minority Oversampling Technique or SMOTE for short. I applied this technique before and I got descent results. I think is a good method and popular one because it looks to balancing the class distribution and take care on creating overfitting of the majority class by creating larger and less specific decision regions.

```python
X_oversample, y_oversample = SMOTE().fit_resample(X_train, y_train)
```

```python
fig, ax = plt.subplots(1,2, figsize = (18,6))
plot_sampling(X_train, y_train, X_oversample, y_oversample, ax, fig, sampler = SMOTE())
```

####  <span style="color:#28BAE5"> Undersample  <a name="id3"></a>

Another aproach worthy to explore is the undersample, even if the dataset is only 30K samples. In this approach, we reduce the number of samples from the majority class to match the number of samples in the minority class. It might performs poorly than the oversampling, but could give me a ranges to evaluate models.

```python
nm2 = NearMiss(version=2)
X_undersample, y_undersample = nm2.fit_resample(X_train, y_train)
```

```python
fig, ax = plt.subplots(1,2, figsize = (18,6))
plot_sampling(X_train, y_train, X_undersample, y_undersample, ax, fig, sampler=NearMiss(version=2))
```

####  <span style="color:#67bf5c"> Models Evaluation  <a name="id3"></a>

```python
%%time
evaluacion = list()
for i, m in clasificadores.items():
    print ('EVALUANDO MODELO: {model}'.format(model=i))
    model = {}
    model['name'] = i
    y_pred_train = m.predict(X_train)
    y_pred_val = m.predict(X_val)
    y_prob_train = m.predict_proba(X_train)
    y_prob_val = m.predict_proba(X_val)
        
    model['accuracy_train'] = accuracy_score(y_true=y_train, y_pred=y_pred_train)
    model['accuracy_vals'] = accuracy_score(y_true=y_val, y_pred=y_pred_val)
    model['precision_train'] = precision_score(y_true=y_train, y_pred=y_pred_train)
    model['precision_vals'] = precision_score(y_true=y_val, y_pred=y_pred_val)
    model['recall_train'] = recall_score(y_true=y_train, y_pred=y_pred_train)
    model['recall_vals'] = recall_score(y_true=y_val, y_pred=y_pred_val)
    model['f1_train'] = f1_score(y_true=y_train, y_pred=y_pred_train)
    model['f1_vals'] = f1_score(y_true=y_val, y_pred=y_pred_val)
    model['AUC_train'] = roc_auc_score(y_train, y_pred_train)
    model['AUC_val'] = roc_auc_score(y_val, y_pred_val)
    evaluacion.append(model)
```

```python
# Contruccion de un DataFrame para visualizar resultados
df = pandas.DataFrame.from_dict(evaluacion)
df.set_index("name", inplace=True)
df.sort_values('AUC_val', ascending=False)
```

```python
%%time
# Calculamos AUC y las matrices de confusiÃ³n
msc = list()
for k, v in clasificadores.items():
    print("\n----------------------------------------------------------------")
    print ('Obteniendo AUC de: {model}'.format(model=k))
    model = {}
    model['name'] = k
    y_pred_val = v.predict(X_val)
    y_pred_train = v.predict(X_train)
    #======================================
    print('\n')
    roc_auc_train = roc_auc_score(y_train, y_pred_train)
    print(f"AUC_train: {roc_auc_train:.2f}")
    roc_auc_val = roc_auc_score(y_val, y_pred_val)
    print(f"AUC_val: {roc_auc_val:.2f}")
    print('\n')
    #======================================
    y_probas_train = v.predict_proba(X_train)
    preds_probtrain = y_probas_train [:,1]
    y_probas_val = v.predict_proba(X_val)
    preds_probval = y_probas_val [:,1]
    #======================================
    sns.set_style('whitegrid')
    sns.set_palette("Set2")
    #print("\n----------------------------------------------------------------")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[14, 8])
    sns.set(font_scale=1.1)
    skplt.metrics.plot_roc_curve(y_val, y_probas_val, ax=ax1)
    skplt.metrics.plot_roc_curve(y_train, y_probas_train, ax=ax2)
    plt.show()
    #======================================
    conf_matrixval = confusion_matrix(y_val, y_pred_val)
    conf_matrixtrain = confusion_matrix(y_train, y_pred_train)
    model['conf_matrixvalnorm'] = (conf_matrixval/(sum(conf_matrixval)))*100
    model['conf_matrixtrainnorm'] = (conf_matrixtrain/(sum(conf_matrixtrain)))*100

    msc.append(model)
```
