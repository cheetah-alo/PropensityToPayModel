####  <span style="color:#a2a2a2"> StratifiedKFold  <a name="id3"></a>

```python
MMS = MinMaxScaler()
X = loadtxt((config_output + "X.csv"), delimiter=',')
y_CV = loadtxt((config_output + "y.csv"), delimiter=',')
```

```python
X_CV = MinMaxScaler().fit_transform(X)
```

#### <span style="color:#a2a2a2"> Optimizacion de Hiperparametros  <a name="id3"></a>

```python
from sklearn.model_selection import StratifiedKFold

folds = StratifiedKFold(6)

def models(classif, X, y, folds, hyperparams = {}):
    
    metrics = []
    
    for i, (train_idx, val_idx) in enumerate(folds.split(X, y)):
        
        X_train, y_train = X[train_idx], y[train_idx]
        X_val, y_val = X[val_idx], y[val_idx]
        
        model = classif(**hyperparams)
        model.fit(X_train, y_train)
        
        y_train = model.predict(X_train)
        y_val = model.predict(X_val)
        
        metrics.append({
            "FOLD": i + 1,
            "Accuracy": accuracy_score(y_train, y_train),
            "ROC AUC" : roc_auc_score(y_train, y_train),
            'f1': f1_score(y_train, y_train),
            "Conf_matrix" : confusion_matrix(y_train, y_train),
            "SET": "TRAIN"
        })

        metrics.append({
            "FOLD": i + 1,
            "Accuracy": accuracy_score(y_val, y_val),
            "ROC AUC" : roc_auc_score(y_val, y_val),
            'f1': f1_score(y_val, y_val),
            "Conf_matrix" : confusion_matrix(y_val, y_val),
            "SET": "VAL"
        })
        
        
    metrics = pandas.DataFrame.from_records(metrics)
   
    fig, ax1 = plt.subplots(figsize=[14, 4])
    sns.barplot(x = "SET", y = "Accuracy", data = metrics.groupby("SET").mean().reset_index(), ax = ax1, alpha = 0.6)
    fig, ax2 = plt.subplots(figsize=[14, 4])
    sns.barplot(x = "FOLD", y = "Accuracy", hue = "SET", data = metrics, ax = ax2, alpha = 0.6)
    plt.show()
  
    fig, ax3 = plt.subplots(figsize=[14, 4])
    sns.barplot(x = "SET", y = "ROC AUC", data = metrics.groupby("SET").mean().reset_index(), ax = ax3, alpha = 0.6)
    fig, ax4 = plt.subplots(figsize=[14, 4])
    sns.barplot(x = "FOLD", y = "ROC AUC", hue = "SET", data = metrics, ax = ax4, alpha = 0.6)
    plt.show()
```

```python
xgbboost = models(XGBClassifier, X_CV, y_CV, folds,
    hyperparams = {'gamma': 0.8,'learning_rate':0.06, 'max_depth':10,'random_state':1,'n_estimators':65, 'n_jobs':0, 'reg_alpha': 0.8, 'reg_lambda': 0.4}
            )
```

```python
%%time
xboboost = XGBClassifier(objective ='reg:logistic', gamma = 0.8, learning_rate = 0.06, max_depth=10, random_state=1, n_estimators=65, n_jobs=0, reg_alpha = 0.8, reg_lambda= 0.4)
xboboost.fit(X_train, y_train)
```

####  <span style="color:#28BAE5"> Diccionario con los clasificadores  <a name="id3"></a>

```python
clasificadores = {
                  'LogisticReg': clrboost,
                  'Random Forest': rfcboost,
                  'XGBClassifier': xboboost,
                 }
```

```python
%%time
evaluacion = list()
for i, m in clasificadores.items():
    print ('EVALUANDO MODELO: {model}'.format(model=i))
    model = {}
    model['name'] = i
    y_pred_train = m.predict(X_train)
    y_pred_val = m.predict(X_val)
    y_prob_train = m.predict_proba(X_train)
    y_prob_val = m.predict_proba(X_val)
        
    model['accuracy_train'] = accuracy_score(y_true=y_train, y_pred=y_pred_train)
    model['accuracy_vals'] = accuracy_score(y_true=y_val, y_pred=y_pred_val)
    model['precision_train'] = precision_score(y_true=y_train, y_pred=y_pred_train)
    model['precision_vals'] = precision_score(y_true=y_val, y_pred=y_pred_val)
    model['recall_train'] = recall_score(y_true=y_train, y_pred=y_pred_train)
    model['recall_vals'] = recall_score(y_true=y_val, y_pred=y_pred_val)
    model['f1_train'] = f1_score(y_true=y_train, y_pred=y_pred_train)
    model['f1_vals'] = f1_score(y_true=y_val, y_pred=y_pred_val)
    model['AUC_train'] = roc_auc_score(y_train, y_pred_train)
    model['AUC_val'] = roc_auc_score(y_val, y_pred_val)
    evaluacion.append(model)
```

####  <span style="color:#67bf5c"> Model Metrics  <a name="id3"></a>

```python
%%time
# Calculamos AUC y las matrices de confusión
msc = list()
for k, v in clasificadores.items():
    print("\n----------------------------------------------------------------")
    print ('Obteniendo AUC de: {model}'.format(model=k))
    model = {}
    model['name'] = k
    y_pred_val = v.predict(X_val)
    y_pred_train = v.predict(X_train)
    #======================================
    print('\n')
    roc_auc_train = roc_auc_score(y_train, y_pred_train)
    print(f"AUC_train: {roc_auc_train:.2f}")
    roc_auc_val = roc_auc_score(y_val, y_pred_val)
    print(f"AUC_val: {roc_auc_val:.2f}")
    print('\n')
    #======================================
    y_probas_train = v.predict_proba(X_train)
    preds_probtrain = y_probas_train [:,1]
    y_probas_val = v.predict_proba(X_val)
    preds_probval = y_probas_val [:,1]
    #======================================
    sns.set_style('whitegrid')
    sns.set_palette("Set2")
    #print("\n----------------------------------------------------------------")
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[14, 8])
    sns.set(font_scale=1.1)
    skplt.metrics.plot_roc_curve(y_val, y_probas_val, ax=ax1)
    skplt.metrics.plot_roc_curve(y_train, y_probas_train, ax=ax2)
    plt.show()
    #======================================
    conf_matrixval = confusion_matrix(y_val, y_pred_val)
    conf_matrixtrain = confusion_matrix(y_train, y_pred_train)
    model['conf_matrixvalnorm'] = (conf_matrixval/(sum(conf_matrixval)))*100
    model['conf_matrixtrainnorm'] = (conf_matrixtrain/(sum(conf_matrixtrain)))*100

    msc.append(model)
```
